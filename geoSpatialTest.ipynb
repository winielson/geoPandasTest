```python
import pandas as pd
import geopandas as gpd
```

### Import geojson file into geopandas dataframe

```python
df = gpd.read_file(r"data\WorldAware_Data.geojson")
df
```

### Display dataframe info
Shows that all columns are filled

```python
print(df.info())
```

### Get geometry of each record

```python
df.geometry
```

### Get area of each polygon

```python
df.geometry.area
```

### Sort by category_name column

```python
df.sort_values(by=['category_name'], inplace=True)
df
```

### Get value counts of category_name column

```python
df.category_name.value_counts()
```

### Get value counts of intelType_name column

```python
df.intelType_name.value_counts()
```

### Only show rows with category_name == 'Wildfire'
(Select rows from a DataFrame based on column values)

```python
df2 = df.loc[df['category_name'] == 'Wildfire']
df2
```

### Export dataframe to geojson file

```python
df2.to_file("wildfires.geojson", driver="GeoJSON")
```
